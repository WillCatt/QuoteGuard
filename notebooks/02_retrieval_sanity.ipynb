# Retrieval Sanity Check

Use this notebook to inspect top-k results for a curated question set after building a real index.

```python
from pathlib import Path
from quoteguard.retrieval.service import RetrievalService
from quoteguard.ingestion.models import Chunk
import json

Path('../data/processed/chunks/chunks.jsonl')
```

```python
chunks_path = Path('../data/processed/chunks/chunks.jsonl')
chunks = [Chunk.model_validate(json.loads(line)) for line in chunks_path.read_text().splitlines() if line.strip()]
retrieval = RetrievalService(chunks, store_path=Path('../data/processed/chroma/vector_store.jsonl'))
retrieval.query('Does the policy cover theft?', k=5, filters={'product_type': 'home_contents'})
```
