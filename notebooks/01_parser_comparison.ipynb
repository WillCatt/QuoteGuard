# Parser Comparison

Use this notebook to compare `pymupdf4llm` and `docling` on a small set of real insurer PDFs. The codebase now exposes explicit parser backends in `quoteguard.ingestion.parser`.

```python
import sys
from pathlib import Path

ROOT_DIR = Path.cwd().resolve().parent
SRC_DIR = ROOT_DIR / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from quoteguard.ingestion.parser import ParserBackend, get_parser

sample_paths = list(Path('../data/raw_pdfs').glob('*.pdf'))[:3]
sample_paths
```

```python
for backend in [ParserBackend.PYMUPDF4LLM, ParserBackend.DOCLING]:
    parser = get_parser(backend)
    print(backend.value, parser.__class__.__name__)
    for path in sample_paths:
        doc = parser.parse(path)
        print(path.name, doc.parser_backend, len(doc.sections))
    print('---')
```
